```python
import pandas as pd
```

```python
# CSV Files
happy_2017_path = "happy_2017.csv"
happy_2018_path = "happy_2018.csv"
happy_2019_path = "happy_2019.csv"
happy_2020_path = "happy_2020.csv"
```

```python
# List of columns
happy_columns = ["City", "State_id", "Emotional & Physical Well-Being Rank", "Income & Employment Rank", "Community & Environment Rank"]
happy_columns
```

```python
# Reading in 2017 file & Renaming 2017 Columns
happy_2017_df = pd.read_csv(happy_2017_path)
happy_2017_df.rename(columns = {'Emotional & Physical Well-Being Rank': 'EPRank2017',
 'Income & Employment Rank': 'IERank2017',
 'Community & Environment Rank': 'CERank2017'}, inplace = True)
happy_2017_df
```

```python
# Reading in 2018 file & Renaming 2018 Columns
happy_2018_df = pd.read_csv(happy_2018_path, usecols = happy_columns)
happy_2018_df.rename(columns = {'Emotional & Physical Well-Being Rank': 'EPRank2018',
 'Income & Employment Rank': 'IERank2018',
 'Community & Environment Rank': 'CERank2018'}, inplace = True)
happy_2018_df
```

```python
# Reading in 2019 file & Renaming 2019 Columns
happy_2019_df = pd.read_csv(happy_2019_path, usecols = happy_columns)
happy_2019_df.rename(columns = {'Emotional & Physical Well-Being Rank': 'EPRank2019',
 'Income & Employment Rank': 'IERank2019',
 'Community & Environment Rank': 'CERank2019'}, inplace = True)
happy_2019_df
```

```python
# Reading in 2020 file & Renaming 2020 Columns
happy_2020_df = pd.read_csv(happy_2020_path, usecols = happy_columns)
happy_2020_df.rename(columns = {'Emotional & Physical Well-Being Rank': 'EPRank2020',
 'Income & Employment Rank': 'IERank2020',
 'Community & Environment Rank': 'CERank2020'}, inplace = True)
happy_2020_df
```

```python
# Merging the dataframes on emotional and physical well-being rank
emotional_df = pd.merge(happy_2017_df, happy_2018_df, on=["City", "State_id"])
emotional_df
```

```python
emotional2_df = pd.merge(emotional_df, happy_2019_df, on=["City", "State_id"])
emotional2_df
```

```python
emotional_physical_df = pd.merge(emotional2_df, happy_2020_df, on=["City", "State_id"])
emotional_physical_df
```

```python
# Export to JSON
emotional_physical_df.head().to_json('Happy.json', orient='records')
```
