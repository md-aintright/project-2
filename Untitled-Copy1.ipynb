```python
import csv
import pandas as pd
```

```python
df1 = pd.read_csv("happy.2017.merged.csv")
df1['year'] = "2017"
df1.head()
```

```python
df2 = pd.read_csv("happy.2018.merged.csv")
df2['year'] = "2018"
df2.head()
```

```python
df3 = pd.read_csv("happy.2019.merged.csv")
df3['year'] = "2019"
df3.head()
```

```python
df4 = pd.read_csv("happy.2020.merged.csv")
df4['year'] = "2020"
df4.head()
```

```python
merged_df1 = pd.DataFrame.merge(df1, df2, how="outer")
merged_df1.head()
```

```python
merged_df2 = pd.DataFrame.merge(df3, df4, how="outer")
merged_df2.head()
```

```python
merged_df = pd.DataFrame.merge(merged_df1, merged_df2, how="outer")
merged_df.head()
```

```python
#export to csv
merged_df.to_csv('out2.csv')
```
